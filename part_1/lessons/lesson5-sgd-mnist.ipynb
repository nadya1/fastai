Now instead we'll use PyTorch's functions to do it for us, and also to handle mini-batches (which we didn't do last time, since our dataset was so small).

````python
bs=64
train_ds = TensorDataset(x_train, y_train)
valid_ds = TensorDataset(x_valid, y_valid)
data = DataBunch.create(train_ds, valid_ds, bs=bs)
````

````python
x,y = next(iter(data.train_dl))
x.shape,y.shape
````

````python
class Mnist_Logistic(nn.Module):
    def __init__(self):
        super().__init__()
        self.lin = nn.Linear(784, 10, bias=True)

    def forward(self, xb): return self.lin(xb)
````

````python
model = Mnist_Logistic().cuda()
````

````python
model
````

````python
model.lin
````

````python
model(x).shape
````

````python
[p.shape for p in model.parameters()]
````

````python
lr=2e-2
````

````python
loss_func = nn.CrossEntropyLoss()
````

````python
def update(x,y,lr):
    wd = 1e-5
    y_hat = model(x)
    # weight decay
    w2 = 0.
    for p in model.parameters(): w2 += (p**2).sum()
    # add to regular loss
    loss = loss_func(y_hat, y) + w2*wd
    loss.backward()
    with torch.no_grad():
        for p in model.parameters():
            p.sub_(lr * p.grad)
            p.grad.zero_()
    return loss.item()
````

````python
losses = [update(x,y,lr) for x,y in data.train_dl]
````

````python
plt.plot(losses);
````

````python
class Mnist_NN(nn.Module):
    def __init__(self):
        super().__init__()
        self.lin1 = nn.Linear(784, 50, bias=True)
        self.lin2 = nn.Linear(50, 10, bias=True)

    def forward(self, xb):
        x = self.lin1(xb)
        x = F.relu(x)
        return self.lin2(x)
````

````python
model = Mnist_NN().cuda()
````

````python
losses = [update(x,y,lr) for x,y in data.train_dl]
````

````python
plt.plot(losses);
````

````python
model = Mnist_NN().cuda()
````

````python
def update(x,y,lr):
    #PyTorch is taking model.paramters and optimace them using "SGD, ADAM, or any optim.xx"
    opt = optim.Adam(model.parameters(), lr) #or use optim.SGD()... 
    y_hat = model(x)
    loss = loss_func(y_hat, y)
    loss.backward()
    opt.step() # this is equivalant to:  for p in model.parameters():   p.sub_(lr * p.grad)
    opt.zero_grad()
    return loss.item()
````

````python
losses = [update(x,y,1e-3) for x,y in data.train_dl]
````

````python
plt.plot(losses);
````

````python
# ERROR: with high Lr
#learning right high: e.g 2e-2
losses = [update(x,y,2e-2) for x,y in data.train_dl]
````

````python
plt.plot(losses);
````

````python
learn = Learner(data, Mnist_NN(), loss_func=loss_func, metrics=accuracy)
````

````python
%debug
````

````python
learn.lr_find()
learn.recorder.plot()
````

````python
learn.fit_one_cycle(1, 1e-2)
````

````python
learn.recorder.plot_lr(show_moms=True)
````

# SGD vs Adam

````python
model = Mnist_NN().cuda()
````

````python
def update(x,y,lr, optim_opt):
    #PyTorch is taking model.paramters and optimace them using "SGD, ADAM, or any optim.xx" 
    if optim_opt == 'ADAM':
        opt = optim.Adam(model.parameters(), lr)   
    elif optim_opt == 'SGD':
        opt = optim.SGD(model.parameters(), lr) 
    elif 'SGD-momt' in optim_opt:
        opt = optim.SGD(model.parameters(), lr, momentum=0.9) 
    y_hat = model(x)
    loss = loss_func(y_hat, y)
    loss.backward()
    opt.step() # this is equivalant to:  for p in model.parameters():   p.sub_(lr * p.grad)
    opt.zero_grad()
    return loss.item()
````

````python
losses_cmp={}
optim_lr = [['ADAM',1e-3],['SGD',2e-2],['SGD-momt',2e-2]]#,['SGD-momt2',1e-3]]
for opt, lr in optim_lr:
    model = Mnist_NN().cuda()
    print ('Getting losses with Opt: %s, Lr:%s\t ....' % (opt,lr)) 
    losses_cmp[opt] = [update(x,y,lr, optim_opt=opt) for x,y in data.train_dl]
````

````python
for title, lr in optim_lr: 
    plt.plot(losses_cmp[title],label= '%s:%s'%(title,lr))
plt.legend(loc='best')
plt.ylabel('Loss')
plt.xlabel('Batches processed')
````
